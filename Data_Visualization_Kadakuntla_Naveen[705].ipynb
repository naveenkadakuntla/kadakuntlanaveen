```python
# Importing the required libraries
import pandas as pd
from matplotlib import pyplot as plt
from urllib.request import urlretrieve
```

```python
# Retrieving the CSV data from the URL

italy_covid_url = 'https://gist.githubusercontent.com/aakashns/f6a004fa20c84fec53262f9a8bfee775/raw/f309558b1cf5103424cef58e2ecb8704dcd4d74c/italy-covid-daywise.csv'

urlretrieve(italy_covid_url, 'italy-covid-daywise.csv')
```

```python
# Reading the CSV File using the Pandas Library

covid_data = pd.read_csv('italy-covid-daywise.csv')
```

```python
covid_data
```

```python
Months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug"]

Cases_Data = [covid_data[1:32].new_cases.sum(), covid_data[32:61].new_cases.sum(), covid_data[61:92].new_cases.sum(), covid_data[92:122].new_cases.sum(), covid_data[122:153].new_cases.sum(), covid_data[153:183].new_cases.sum(), covid_data[183:214].new_cases.sum(), covid_data[214:245].new_cases.sum()]

Deaths_Data = [covid_data[1:32].new_deaths.sum(), covid_data[32:61].new_deaths.sum(), covid_data[61:92].new_deaths.sum(), covid_data[92:122].new_deaths.sum(), covid_data[122:153].new_deaths.sum(), covid_data[153:183].new_deaths.sum(), covid_data[183:214].new_deaths.sum(), covid_data[214:245].new_deaths.sum()]

Tests_Data = [covid_data[1:32].new_tests.sum(), covid_data[32:61].new_tests.sum(), covid_data[61:92].new_tests.sum(), covid_data[92:122].new_tests.sum(), covid_data[122:153].new_tests.sum(), covid_data[153:183].new_tests.sum(), covid_data[183:214].new_tests.sum(), covid_data[214:245].new_tests.sum()]
```

```python
plt.bar(Months, Cases_Data, label = "cases")
plt.bar(Months, Deaths_Data, label = "deaths")
plt.title("Italy's Covid Data 2020 (Jan - Aug)")
plt.xlabel("MONTHS")
plt.ylabel("NUMBER OF REGISTERED")
plt.legend()
```

```python
plt.pie(Deaths_Data, labels = Months)
plt.legend(title = "Monthly Deaths")
```

```python
plt.plot(covid_data.date, covid_data.new_cases, label = "Cases")
plt.plot(covid_data.date, covid_data.new_tests, label = "Tests")
plt.title("Italy Covid Data Jan - Aug")
plt.xlabel("DATE")
plt.ylabel("NUMBER OF REGISTERED")
plt.legend()
```

```python
plt.scatter(covid_data[61:92].date, covid_data[61:92].new_deaths)
plt.title("Italy Covid-19 Deaths in March 2020")
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.ylabel("Number of deaths")
```
